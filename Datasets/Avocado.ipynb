```python
import pandas as pd
df = pd.read_csv("avocado.csv")
```

```python
df = df.copy()[df["type"] == "organic"]
df['Date'] = pd.to_datetime(df["Date"])
df_albany = df.copy()[df["region"] == "Albany"]
df_albany.set_index("Date", inplace = True)
df_albany.sort_index(inplace = True)
df_albany["AveragePrice"].rolling(1).mean().dropna().plot()
```

```python
df["region"].unique()
graph_df = pd.DataFrame()
for region in df["region"].unique():
    print(region)
    region_df = df.copy()[df["region"] == region]
    region_df.set_index("Date", inplace = True)
    region_df.sort_index(inplace = True)
    region_df[f'{region}25ma'] = region_df["AveragePrice"].rolling(25).mean().dropna()
    if graph_df.empty:
        graph_df = region_df[[f'{region}25ma']]
    else:
        graph_df = graph_df.join(region_df[f'{region}25ma'])
```

```python
graph_df.plot(legend = False)
```
